Simple notebook to visualise the partial pointcloud generated in `scripts/shape_completion.py`

```python
import numpy as np
import results.runs_sdf as runs_sdf
import os
import plotly.graph_objects as go
```

```python
path = os.path.join(os.path.dirname(runs_sdf.__file__), '17_07_172540', 'infer_latent_19_07_114157', 'partial_pointcloud.npy')
pointcloud = np.load(path)
```

```python
fig = go.Figure(
    [
        go.Scatter3d(x=pointcloud[:, 0], 
                     y=pointcloud[:, 1],
                     z=pointcloud[:, 2], mode='markers',
        marker=dict(size=1))
    ]
)
fig.update_layout(scene=dict(
                    xaxis=dict(showgrid=False, showbackground=False, showticklabels=False, visible=False),
                    yaxis=dict(showgrid=False, showbackground=False, showticklabels=False, visible=False),
                    zaxis=dict(showgrid=False, showbackground=False, showticklabels=False, visible=False)
                    ),
                )
fig.show()
```
